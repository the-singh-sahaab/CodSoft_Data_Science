**making the formate uniform**


change the dictionary to list

```python
#we will maek a helper function to convert
# here will will use ast.literal_eval() to convert str enclosed dict. in to dict.
import ast
def convert(obj):
  l = []
  for i in ast.literal_eval(obj):
    l.append(i["name"])
  return l
# helping fumction to convert cast str dictionary to the list by only taking first two value
def convert3(obj):
  l = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      l.append(i["name"])
      counter += 1
    else:
      break
  return l
# making a function to fetch director
def bring(obj):
  l = []
  for i in ast.literal_eval(obj):
    if i["job"] == "Director":
      l.append(i["name"])
      break
  return l
```

```python
#converting the genres dictionary into the list
movies_credit['genres'] = movies_credit['genres'].apply(convert)
```

```python
#try to change all the column and replacing it to the orginal place.
#"movie_id", "title", "overview", "genres", 'keywords', 'cast', 'crew
movies_credit['keywords']=movies_credit['keywords'].apply(convert)
movies_credit['cast'] = movies_credit['cast'].apply(convert3)
```

```python
movies_credit['crew'] = movies_credit['crew'].apply(bring)
```

```python
#changing overview column on list from string
movies_credit["overview"] = movies_credit["overview"].apply(lambda x:x.split())
```

```python
movies_credit
```

```python
#removing spaces from every element in columns.
movies_credit["genres"] = movies_credit["genres"].apply(lambda x:[i.replace("","") for i in x])
movies_credit["keywords"] = movies_credit["keywords"].apply(lambda x:[i.replace("","") for i in x])
movies_credit["cast"] = movies_credit["cast"].apply(lambda x:[i.replace("","") for i in x])
movies_credit["crew"] = movies_credit["crew"].apply(lambda x:[i.replace("","") for i in x])
```

**combining all the 4 cloumns**

```python
movies_credit["tag_"] = movies_credit["genres"] + movies_credit["keywords"] + movies_credit["cast"] + movies_credit["crew"]
```

```python
new_m_c = movies_credit[["movie_id", "title","tag_"]]
```

```python
new_m_c["tag_"] = new_m_c["tag_"].apply(lambda x: " ".join(x))
```

```python
!pip install nltk
```

using porter stemmer method for removing the words of same type (eg, loved , loving, love)

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#making a helping function
def Stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_m_c["tag_"] = new_m_c["tag_"].apply(Stem)
```

```python
new_m_c["tag_"][0]
```

starting the process of bag of word . where we are trying to make a mage tag to get the vector for every tag.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
vector_tag = cv.fit_transform(new_m_c["tag_"]).toarray()
vector_tag
"""
now here the vector_tag is having 5000 columns and 4806 row ( means 5000 word were taken form the hash map from the mage-tag (new_m_c) and the 4806 is number of movies)
this show that we have to made 4806 vector and have to plot and have to find the cosin-distance (this is the angle between 2 vector).
 WELL WE WILL BE WORKING WITH THE 'COSINE-SIMILARITY' {it like opposite of cosine-distance}
"""
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector_tag)
```

```python
similarity[1]
```

making the function which will take a movie name and it will return 5 movie name similar to it

```python
def recommd(movie):
  movie_index = new_m_c[new_m_c['title'] == movie].index[0]
  distances = similarity[movie_index]
  sorted_distance = sorted(list(enumerate(distances)),reverse= True ,key = lambda x:x[1])[1:6]
  for i in sorted_distance:
    print(new_m_c.iloc[i[0]].title)
```

```python
recommd("Batman Begins")
```
