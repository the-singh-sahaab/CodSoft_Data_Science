```python
''' ok so this is a single model for credit card fraud detec.
here in this model i have used the random forest model for computing
'''
# =================================================================

# these are the imported lib.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, MaxPooling1D, Flatten
from sklearn.model_selection import train_test_split
# import Randomforestclassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# i have called the file from its path
df = pd.read_csv("creditcard.csv")
df.head()
```

```python
# cleaning the dataset
df.isnull().sum()# this line of code tell how many null values are there in a colunm
```

```python
# here we are removing the null values
df = df.dropna()
df.isnull().sum()

'''notes
the random forest classifier can be used for both classification and regression modeling
thus we have used it here well the special thing about the rfc model is that it doesnt require and kind of data cleaning
still i have clean the data becoz to show it doen in this way.
'''
```

```python
# here we will be using RFM which dont require standardise the data
scaler  = StandardScaler()
```

```python
df['Normailzed_amounnt'] = scaler.fit_transform(df['Amount'].values.reshape(-1,1))
```

```python
df.drop(['Amount', 'Time'], axis = 1, inplace = True)
```

```python
# making all the independent valrible
x = df.drop(["Class"], axis = 1)
y = df["Class"]
```

```python
# making the trian and test data by dividing the data with the testsize of 0.3
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 1)
```

Appling the RandomForest method to Classify the Fraud or non Fraud

```python
rf = RandomForestClassifier (n_estimators = 100)#giving the number of trees in the forest
rf.fit(x_train, y_train)#train our training data
pre_f = rf.predict(x_test)#doing the prediction
rf_froest = rf.score(x_test, y_test)*100
rf_froest#calcuating the accuracy
all = classification_report(y_test, pre_f)
```

```python
print(all)
```

Appling Hybrid machine learning model, where we are using CNN and RandomForest to do the same.

```python
# Create the CNN model
cnn_model = Sequential()
# cnn_model = Sequential()
cnn_model.add(Conv1D(32, kernel_size=3, activation='relu', input_shape=(x_train.shape[1],1)))
cnn_model.add(MaxPooling1D(pool_size=2))
cnn_model.add(Flatten())
cnn_model.add(Dense(128, activation='relu'))
cnn_model.add(Dense(1, activation='sigmoid'))
# Create the random forest model
rf_model = RandomForestClassifier(n_estimators=100)

# Train the CNN model
cnn_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
cnn_model.fit(x_train, y_train, epochs=10)

# Extract the features from the CNN model
cnn_features = cnn_model.predict(x_test)

# Train the random forest model on the CNN features
rf_model.fit(cnn_features, y_test)

# Make predictions on the test set
y_pred = rf_model.predict(cnn_model.predict(x_test))
accuracy = np.sum(y_pred == y_test) / len(y_test)#precision
print('Accuracy:', accuracy*100)
```

```python
a = classification_report(y_test, y_pred)
print(a)
```
